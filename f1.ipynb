```python
import warnings
warnings.filterwarnings('ignore')

#set up a cache
fastf1.Cache.enable_cache(CACHE)  
fastf1.plotting.setup_mpl()

#plot the lap
def plot_lap(lap, driver):
    fig, ax = plt.subplots()
    ax.plot(lap['Time'], lap['Speed'], label='Fast')
    ax.set_xlabel('Time')
    ax.set_ylabel('Speed [Km/h]')
    ax.set_title(f'{driver} lap')
    ax.legend()
    plt.show()
```

```python

session = fastf1.get_session(2022, 'Monza', 'Q')
session.load()

drivers = ['HAM','LEC','VER','RUS']


for driver in drivers:
    lap = session.laps.pick_driver(driver).pick_fastest()
    data = lap.get_car_data()
    plot_lap(data, driver)
```
